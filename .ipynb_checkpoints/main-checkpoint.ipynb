```python
import os
import matplotlib.pyplot as plt
from imageio import imread
from intellegent_placer_lib.intellegent_placer import check_for_place
```

Запустим алгоритм на датасете и выведем:

```python
folder_path = "dataset"
for image_path in os.listdir(folder_path):
    img, polygon, items, result, res_img = check_for_place(folder_path, image_path)
    l = 3 + len(items)
    fig, ax = plt.subplots(1, l, figsize=(22, 22))
    ax[0].imshow(img)
    ax[0].set_title("initial photo")
    ax[1].imshow(polygon)
    ax[1].set_title("polygon")
    for i in range(0, len(items)):
        ax[i+2].imshow(items[i])
        ax[i+2].set_title(f"item num {i}")
    print(f"Result for {image_path} : {result}")
    if result:
        ax[len(items) + 2].imshow(res_img)
    else:
        ax[len(items) + 2].imshow(imread("false.jpg"))
```

На всех примерах из датасета алгоритм отработал корректно, но расположение объектов на некоторых примерах не является оптимальным, в виду того, что алгоритм выбирает первое подходящее место и распологает объекты по порядку, начиная с наибольшего.
